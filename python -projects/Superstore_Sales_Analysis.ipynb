```python
# Checking and dropping duplicates
df.duplicated().sum()
df.drop_duplicates
```

```python
# Converting order and ship Dates to Datetime

df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])

# Addition of new columns
df['Order Year']=df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
df['Order Month Name']=df['Order Date'].dt.strftime('%b')
df['Delivery Days']=(df['Ship Date']-df['Order Date']).dt.days

df.shape
print(df[['Order Date','Order Year','Order Month','Delivery Days']].head())
```

```python
# BUSINESS QUESTIONS

# 1 TOTAL REVENUE PROFIT AND ORDERS
total_sales=df.Sales.sum()
total_profit=df.Profit.sum()
total_orders=df['Order ID'].nunique()
profit_margin=(total_profit/total_sales)*100

print(f'\n Q1 Business Overview:')
print(f' Total Sales:{total_sales:,.2f}')
print(f' Total Profit:{total_profit:,.2f}')
print(f' Profit Margin:{profit_margin:,.1f}%')
print(f' Total Orders:{total_orders}')

```

```python
# Least profitable
least_profitable=df.groupby('Product Name')['Profit'].sum().sort_values().head(1)
least_profitable
```

```python
# 2 Sales by Category
cat_sales=df.groupby("Category")["Sales"].sum().sort_values(ascending=False).round(2)
for cat in cat_sales.items():
    print( cat)
```

```python
df.columns.tolist()
```

```python
# Most profitable products
print(f'\n- Top 5 most profitable products:')
top_products=df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(5)
for product,profit in  top_products.items():
    print(f' {product[:40]}:${profit:,.2f}')
```

```python
# profit by Region
region_sales=df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
for Region in region_sales.items():
       print( Region)
    
```

```python
# Monthly sales trend
monthly=df[df['Order Year']==df['Order Year'].max()].groupby('Order Month')['Sales'].sum()
print(monthly)
```

```python
# monthly sales trend

fig, axes=plt.subplots(1,3, figsize=(18,6))
fig.suptitle("Superstore Sales Dashboard",fontsize=20,fontweight="bold")



# profit by region
region_profit=df.groupby('Region')['Profit'].sum().sort_values()
bar_colors=['#e74c3c'if x < 0 else  '#2ecc71' for x in region_profit.values]
axes[1].bar(region_profit.index,region_profit.values,color=bar_colors)
axes[1].set_title('Total Profit by Region', fontsize=14)
axes[1].set_ylabel('Profit($)')
axes[1].axhline(0, color='black',linewidth=0.8)



# SALES BY CATEGORY
cat_sales=df.groupby("Category",)["Sales"].sum().sort_values()
colors= ['#34989b', '#e74c3c','#2ecc71']
axes[0].barh(cat_sales.index,cat_sales.values,color=colors)
axes[0].set_title('Total sales by category', fontsize=14)
axes[0].set_xlabel('Sales($)')
for i, v in enumerate(cat_sales.values):
    axes[0].text(v +500, i, f'${v:,.0f}',va='center',fontsize=10)


# monthly sales trend
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Order Year']=df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
monthly=df.groupby(['Order Year','Order Month'])['Sales'].sum().reset_index()
latest_year=monthly[monthly['Order Year']==monthly['Order Year'].max()]
month_names=['Jan' ,'Feb', 'March', 'April', 'May' ,'June', 'July', 'August', 'September', 'October', 'November' ,'December']
axes[2].plot(latest_year['Order Month'],
             latest_year['Sales'],
             color='#9b59b6',linewidth=2.5,marker='o',markersize=7)

axes[2].set_title('Monthly Sales Trend ', fontsize=14)
axes[2].set_ylabel('Sales($)')
axes[2].set_xlabel('Month')
axes[2].set_xticks(range(1,13))
axes[2].set_xticklabels(month_names, rotation=45)

plt.tight_layout()
plt.savefig('superstore_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print('Dashboard saved as superstore_dashboard.png')



```
