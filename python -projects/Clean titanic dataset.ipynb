```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
```

```python
print(df.isnull().sum())
df
```

```python
# MISSING VALUES AS PERCENTAGES
missing_values_pct=(df.isnull().sum()/len(df))*100
print(missing_values_pct.round(1))
```

```python
# DROPPING CABIN
drop_cabin=df.drop(columns=['Cabin'], inplace=True)
print(drop_cabin)
```

```python
print(df.columns.tolist())
```

```python
# IMPUTATION
median_age=df['Age'].median()
(df['Age'].fillna(median_age, inplace=True))
print(df['Age'].isnull().sum())
```

```python
# REMOVING DUPLICATES
print("Duplicates_before", df.duplicated().sum())
df.drop_duplicates(inplace=True)

print("Duplicates after:", df.duplicated().sum())
print(df.shape)
```

```python
# FIX DATA TYPES
print(df.dtypes)
```

```python
df['Survived']=df['Survived'].astype('category')
df['Pclass']=df['Pclass'].astype('category')
df['Sex']=df['Sex'].astype('category')
df['Embarked']=df['Embarked'].astype('category')
df.dtypes
```

```python
# RENAMING COLUMNS
df.rename(columns={
    'Pclass':'passenger_class',
    'Sibsp':'siblings_spouses',
    'Parch':'parents_children',
    'Survived':'survived'},
    inplace=True)
df.columns.tolist()
```

```python
# DROPPING IRRELEVANT COLUMNS
df.drop(columns=['PassengerId','Name','Ticket'], inplace=True)
df.shape
```

```python
# DETECT AND HANDLE OUTLIERS
print(df[['Age','Fare']].describe())
```

```python
# FINAL VERIFICATION AFTER CLEANING
df.shape
```

```python
print(df.isnull().sum().sum())
```

```python
df.duplicated().sum()
```

```python
df.to_csv("titanic_cleaned.csv" ,index=False)
print('clean file saved!')
```
